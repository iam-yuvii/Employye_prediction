# Data Dictionary

| Column Name | Description |
|-------------|-------------|
| Age | Age of the employee in years. |
| Attrition | Whether the employee left the company (Yes = Left, No = Stayed). |
| BusinessTravel | How frequently the employee travels for work. |
| Department | Department where the employee works. |
| DistanceFromHome | Distance (in miles) between the employee's home and workplace. |
| Education | Employee's highest education level (coded numerically). |
| Gender | Employee's gender. |
| JobRole | Employee's job title or role. |
| JobSatisfaction | Satisfaction score given by the employee (1 = Low, 4 = High). |
| MonthlyIncome | Employee's monthly salary. |
| OverTime | Whether the employee regularly works overtime. |
| TotalWorkingYears | Total years of professional work experience. |
| WorkLifeBalance | Employee's rating of work-life balance (1 = Poor, 4 = Excellent). |
| YearsAtCompany | Number of years the employee has worked at the current company. |
| YearsSinceLastPromotion | Years since the employee's last promotion. |

```python
atrri = raw_data["Attrition"].copy()
```

```python
i = 0
j = 0 

for value in atrri:
    if value == "Yes":
        i = i + 1
    else:
        j = j + 1 

```

```python
data = [i , j ] 
labels = ["Left Company", "Stayed"]
colors = sns.color_palette('pastel')

fig, ax = plt.subplots()

wedges, texts, autotexts = ax.pie(
    data,
    labels=labels,
    colors=colors,
    startangle=90,
    autopct='%.0f%%',
    wedgeprops=dict(width=0.5)
)
plt.title("Distribution of Employee Attrition", fontsize=14, fontweight="bold")
ax.legend(wedges, labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

**Insight:**
- Most employees (84%) did not leave the company.
- Only 16% of employees left (Attrition = Yes).
- Although the imbalance is not severe, it is significant enough to consider techniques such as stratified train-test splitting and evaluating metrics like Precision, Recall, F1-score, and ROC-AUC instead of relying solely on accuracy.
