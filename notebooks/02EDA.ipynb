```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv("raw_data.csv")
```

```python
bar_data =  raw_data[["Age" , "MonthlyIncome",  "YearsAtCompany" , "DistanceFromHome" , "TotalWorkingYears" ]] 
```

```python
hist= bar_data.hist(bins=5, figsize = (10,10))
```

```python
a = raw_data['Department'].value_counts()
b = raw_data['JobRole'].value_counts()
c= raw_data['BusinessTravel'].value_counts()
d= raw_data['MaritalStatus'].value_counts()
e = raw_data['OverTime'].value_counts()

```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

a.plot.bar(ax=axes[0], title="Department", rot=0)
b.plot.bar(ax=axes[1], title="JobRole", rot=45)
c.plot.bar(ax=axes[2], title="BusinessTravel", rot=0)
d.plot.bar(ax=axes[3], title="MaritalStatus", rot=0)
e.plot.bar(ax=axes[4], title="OverTime", rot=0)

# Remove the unused subplot
fig.delaxes(axes[5])

plt.suptitle("Distribution of Categorical Features", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```
